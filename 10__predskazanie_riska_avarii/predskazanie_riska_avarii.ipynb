Заказчику нужно знать, сколько аварий происходит по месяцам (за весь период).

```python
#sql запрос
query = '''
SELECT EXTRACT(YEAR FROM collision_date) AS year,
          EXTRACT(MONTH FROM collision_date) AS month,
          COUNT(case_id) AS accident_count
    FROM collisions
    GROUP BY year, month
    ORDER BY year, month;
'''
df = pd.read_sql(query, engine)
df[['month', 'year']] = df[['month', 'year']].apply(lambda x: x.astype(int))
df
palette = sns.color_palette("tab10", n_colors=df['year'].nunique())

#Построение графика
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='month', y='accident_count', hue='year', marker='o', palette=palette)

plt.xlabel('Месяц')
plt.ylabel('Количество аварий')
plt.title('Количество аварий по месяцам и годам')
plt.xticks(ticks=range(1, 13), labels=['Январь', 'Февраль', 'Март',
                                       'Апрель', 'Май', 'Июнь',
                                       'Июль', 'Август', 'Сентябрь',
                                       'Октябрь', 'Ноябрь', 'Декабрь'])
plt.legend(title='Год')
plt.grid(axis='y')

plt.show()
```

Больше всего аварий происходит в марте. После мая в 2012 количество аварий резко падает почти на 30000 аварий в месяц. Даных за 2013 и 2020 года просто нет либо же очень мало. В остальные же года (2009, 10,11) данные по проишествиям расспределены примерно равномерно.

Метрику для оценки моделей возьму ROC-AUC. Эту метрику выбрал по следующим причинам:
<br>● ROC-AUC измеряет порядок предсказаний, а не их абсолютные значения. Это помогает сравнивать качество моделей, которые могут выдавать разные диапазоны предсказанных вероятностей.

### <b>DecisionTreeClassifier</b>
